```python
from datetime import date, time
import xlsxwriter
```

```python
workbook = xlsxwriter.Workbook('cell_metadata.xlsx')
worksheet = workbook.add_worksheet()
```

```python
# Add a format for the header cells.
header_format1 = workbook.add_format({
    'border': 1,
    'bg_color': '#20B2AA',
    'bold': True,
    'text_wrap': True,
    'valign': 'vcenter',
    'indent': 1,
})

header_format2 = workbook.add_format({
    'border': 2,
    'bg_color': '#EE82EE',
    'bold': True,
    'text_wrap': True,
    'valign': 'vcenter',
    'indent': 1,
})
```

```python
# Set up layout of the worksheet.
worksheet.set_column('A:A', 68)
worksheet.set_column('B:B', 15)
worksheet.set_column('C:C', 15)
worksheet.set_column('D:D', 15)
worksheet.set_column('E:E', 15)
worksheet.set_column('F:F', 15)
worksheet.set_column('G:G', 15)
worksheet.set_column('H:H', 15)
worksheet.set_column('I:I', 15)
worksheet.set_row(0, 36)
```

```python
# Write the header cells and some data that will be used in the examples.
headingA1 = 'Raw Data File Name'
headingB1 = 'Cell ID'
headingC1 = 'Positive Electrode'
headingD1 = 'Negative Electrode'
headingE1 = 'Seperator'
headingF1 = 'Electrolyte'
headingG1 = 'Project Name'
headingH1 = 'Project Description'

headingA2 = 'Entry Type:'
headingB2 = 'Enter Integer'
headingC2 = 'Dropdown: if not present, input new type (see doc x)'
headingD2 = 'Dropdown'
headingE2 = 'Dropdown'
headingF2 = 'Dropdown'
headingG2 = 'Dropdown'
headingH2 = 'Dropdown'

worksheet.write('A1', headingA1, header_format1)
worksheet.write('B1', headingB1, header_format1)
worksheet.write('C1', headingC1, header_format1)
worksheet.write('D1', headingD1, header_format1)
worksheet.write('E1', headingE1, header_format1)
worksheet.write('F1', headingF1, header_format1)
worksheet.write('G1', headingG1, header_format1)
worksheet.write('H1', headingG1, header_format1)

worksheet.write('A2', headingA2, header_format2)
worksheet.write('B2', headingB2, header_format2)
worksheet.write('C2', headingC2, header_format2)
worksheet.write('D2', headingD2, header_format2)
worksheet.write('E2', headingE2, header_format2)
worksheet.write('F2', headingF2, header_format2)
worksheet.write('G2', headingG2, header_format2)
worksheet.write('H2', headingG2, header_format2)
```

```python
worksheet.data_validation('B2:B500', {'validate': 'integer',
                                   'criteria': '>', 'value': 0})

worksheet.data_validation('C2:C500', {'validate': 'list',
                                  'source': ['x', 'y', 'z', 'proprietary', 'unknown']})

worksheet.data_validation('D2:D500', {'validate': 'list',
                                  'source': ['x', 'y', 'z','proprietary', 'unknown']})

worksheet.data_validation('E2:E500', {'validate': 'list',
                                  'source': ['x', 'y', 'z','proprietary', 'unknown']})

worksheet.data_validation('F2:F500', {'validate': 'list',
                                  'source': ['e1', 'e2', 'e3','proprietary', 'unknown']})

worksheet.data_validation('G2:G500', {'validate': 'list',
                                  'source': ['x', 'y', 'z','proprietary', 'unknown']})

worksheet.data_validation('H2:H500', {'validate': 'list',
                                  'source': ['x', 'y', 'z','proprietary', 'unknown']})
```
